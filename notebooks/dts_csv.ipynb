### DTS csv format

```python
# Import necessary dependencies
import datetime, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load DTS data
file_path = '../data/dts/temp_cal_valid_cable_rmnoise.csv'
```

```python
# Read CSV into DataFrame and parse datetime columns
df_temp = pd.read_csv(file_path, index_col=0)
df_temp.columns = pd.to_datetime(df_temp.columns)  # Convert column headers to datetime
df_temp.head()
```

```python
# Transpose to have time as the first axis (axis=0)
temp_arr = df_temp.to_numpy().T  
times = pd.to_datetime(df_temp.columns)  # Time vector
dists = df_temp.index.to_numpy()         # Distance vector (optical fiber length)
```

```python
# Use optical distances corresponding to East Ridge
start_dist = 2820  # meters
end_dist = 3150    # meters
```

```python
# Define time window of interest
start_time = datetime.datetime(2023, 8, 11)
end_time = datetime.datetime(2023, 8, 12)
```

```python
# Find nearest indices for slicing
t_idx_start = np.argmin(np.abs(times - start_time))
t_idx_end = np.argmin(np.abs(times - end_time))
d_idx_start = np.argmin(np.abs(dists - start_dist))
d_idx_end = np.argmin(np.abs(dists - end_dist))
```

```python
# Slice arrays for plotting
plot_arr = temp_arr[t_idx_start:t_idx_end, d_idx_start:d_idx_end]
plot_times = times[t_idx_start:t_idx_end]
plot_dists = dists[d_idx_start:d_idx_end]
```

```python
fig, ax = plt.subplots(figsize=(16, 7))

im1 = ax.imshow(
    plot_arr.T,          # Transpose to plot distance vs time
    origin='lower',      # Keep origin bottom-left
    aspect='auto',       # Auto-scale axis ratio
    extent=[
        plot_times[0], 
        plot_times[-1], 
        plot_dists[0], 
        plot_dists[-1]
    ],
    cmap='plasma',                
    vmin=np.percentile(plot_arr[~np.isnan(plot_arr)], 1),   
    vmax=np.percentile(plot_arr[~np.isnan(plot_arr)], 99),  
)

ax.set_xlabel('Time')
ax.set_ylabel('Optical Distance along Cable [m]')
ax.invert_yaxis()

cbar = fig.colorbar(im1, pad=0.01)
cbar.set_label('Temperature [Â°C]')

plt.tight_layout()
plt.show()
```
